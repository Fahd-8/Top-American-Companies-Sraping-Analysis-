```python
"""**Web scraping**"""

import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python

world_titles = table.find_all('th')

world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
df
```

```python
# saving the csv file

df.to_csv('Companies.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Companies.csv')

# Select the first 40 rows
df_subset = df.head(40)

# Plot Rank against Employees for the subset
plt.figure(figsize=(13, 8))

# Plot with customizations
plt.plot(df_subset['Rank'], df_subset['Employees'], marker='o', linestyle='-', color='yellow', linewidth=2, markersize=8, label='Employees')
plt.title('Rank vs Employees', fontsize=16)
plt.xlabel('Rank', fontsize=14)
plt.ylabel('Employees', fontsize=14)
plt.xticks(range(1, 41))
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
